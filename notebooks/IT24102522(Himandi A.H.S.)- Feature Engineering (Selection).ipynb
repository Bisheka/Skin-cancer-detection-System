```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, f_classif

#Load the dataset
df = pd.read_csv('D:\\SLIIT\\Y2S1\\AI & ML\\Project\\2025-Y2-S1-MLB-B9G2-05\\cleaned_metadata_with_image_paths.csv')
```

```python
# Separate features and target
X = df[['age', 'sex']]
y = df['dx']

# Convert 'sex' to numerical for feature selection
X = pd.get_dummies(X)

# Apply SelectKBest
selector = SelectKBest(score_func=f_classif, k=2)
selector.fit(X, y)

# Get selected feature scores
print("Feature scores:", selector.scores_)
```

```python
# Prepare features and target
X = pd.get_dummies(df[['age', 'sex']])
y = df['dx']

# Apply SelectKBest
selector = SelectKBest(score_func=f_classif, k=2)
selector.fit(X, y)

# Create a bar graph of feature importance
plt.figure(figsize=(8, 5))
plt.bar(X.columns, selector.scores_, color='salmon')
plt.title("Feature Importance Scores")
plt.xlabel("Features")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()
```
