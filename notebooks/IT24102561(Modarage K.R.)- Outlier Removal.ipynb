```python
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
#Load the dataset
df = pd.read_csv('D:\\SLIIT\\Y2S1\\AI & ML\\Project\\2025-Y2-S1-MLB-B9G2-05\\cleaned_metadata_with_image_paths.csv')
```

```python

# Member 3: Modarage K.R. - Outlier Removal
# Calculate Q1, Q3, and IQR for 'age'
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]

# Display the shape after outlier removal
print("Shape after outlier removal:", df.shape)
```

```python
# Remove outliers using IQR method
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]

# Create a box plot for 'age'
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['age'], palette='pastel')
plt.title("Box Plot of Age after Outlier Removal")
plt.ylabel("Age")
plt.show()
```
